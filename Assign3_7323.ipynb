```python
import pandas as pd
import numpy as np
```

```python
#Loading dataset

df = pd.read_csv("Iris.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#part1

#Group Numeric Variable by Species
grouped_stats = df.groupby("Species")["PetalLengthCm"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Standard_Deviation="std"
)

grouped_stats
```

```python
#Summary Statistics for All Numeric Variables


all_numeric_stats = df.groupby("Species").agg(
    ["mean", "median", "min", "max", "std"]
)

all_numeric_stats
```

```python
#Create List Containing Numeric Values for Each Category

petal_length_by_species = {}

for species in df["Species"].unique():
    petal_length_by_species[species] = df[df["Species"] == species]["PetalLengthCm"].tolist()

petal_length_by_species
```

```python
species_list = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
#Mean, Standard Deviation, and Percentiles

for species in species_list:
    print("\n====================================")
    print(f"Statistical Details for {species}")
    print("====================================")
    
    species_data = df[df['Species'] == species]
    numeric_columns = species_data.select_dtypes(include=np.number)
    
    print("\nMean:")
    print(numeric_columns.mean())
    
    print("\nStandard Deviation:")
    print(numeric_columns.std())
    
    print("\nPercentiles (25%, 50%, 75%):")
    print(numeric_columns.quantile([0.25, 0.50, 0.75]))
```
