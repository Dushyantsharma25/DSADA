```python
import pandas as pd
import numpy as np
```

```python
#Loading dataset

df = pd.read_csv("shopping_data.csv")
df.head()
```

```python
df.info()
```

```python
#Group Numeric Variable by Gender
grouped_stats = df.groupby("Genre")["Annual Income (k$)"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Standard_Deviation="std"
)

grouped_stats
```

```python
#Summary Statistics for All Numeric Variables


all_numeric_stats = df.groupby("Genre").agg(
    ["mean", "median", "min", "max", "std"]
)

all_numeric_stats
```

```python
#Create List Containing Numeric Values for Each Category

petal_length_by_species = {}

for species in df["Genre"].unique():
    petal_length_by_species[species] = df[df["Genre"] == species]["Annual Income (k$)"].tolist()

petal_length_by_species
```

```python
species_list = ['Male','Female']
```

```python
#Mean, Standard Deviation, and Percentiles

for species in species_list:
   
    print(f"Statistical Details for {species}")
   
    
    species_data = df[df['Genre'] == species]
    numeric_columns = species_data.select_dtypes(include=np.number)
    
    print("\nMean:")
    print(numeric_columns.mean())
    
    print("\nStandard Deviation:")
    print(numeric_columns.std())
    
    print("\nPercentiles (25%, 50%, 75%):")
    print(numeric_columns.quantile([0.25, 0.50, 0.75]))
```
